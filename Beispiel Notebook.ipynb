```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, save, output_notebook, reset_output
from bokeh.models import ColumnDataSource, HoverTool, Plot, LinearAxis, Grid
from bokeh.palettes import Category20
from bokeh.io import show
from bokeh.core.properties import value
output_notebook()
```

```python
#Einlesen
corona=pd.read_csv('Corona Daten.csv',delimiter=',')
corona
```

```python
sum_cases=corona.groupby("countryterritoryCode").sum()
top_cases=sum_cases.nlargest(10,columns="cases")

corona_source=ColumnDataSource(top_cases)
top_cases
```

```python
coronacases_plot=figure(width=800, height=300, y_range=corona_source.data['countryterritoryCode'])
coronacases_plot.hbar(right='cases',y='countryterritoryCode', height=0.8, source=corona_source)
show(coronacases_plot)
```

```python
coronadeath_plot=figure(width=800, height=300, y_range=corona_source.data['countryterritoryCode'])
coronadeath_plot.hbar(right='deaths',y='countryterritoryCode', height=0.8, source=corona_source)
show(coronadeath_plot)
```
